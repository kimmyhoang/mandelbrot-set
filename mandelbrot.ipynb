```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandelbrot(realNum, imagNum, max_iter):
    c = complex(realNum, imagNum)
    z = 0.0j
    
    for i in range(max_iter):
        z = z*z + c
        if(z.real*z.real + z.imag*z.imag) >= 4:
            return i
        
    return max_iter
```

```python
columns = 2000
rows = 2000

result = np.zeros([rows, columns])
for row_index, realNum in enumerate(np.linspace(-2, 1, rows)):
    for column_index, imagNum in enumerate(np.linspace(-1, 1, columns)):
        result[row_index, column_index] = mandelbrot(realNum, imagNum, 100)
    
plt.figure(dpi=100)
plt.imshow(result.T, cmap='gist_ncar', interpolation='bilinear',extent=[-2, 1, -1, 1])
plt.xlabel('Re')
plt.ylabel('Im')
plt.show()    
```
